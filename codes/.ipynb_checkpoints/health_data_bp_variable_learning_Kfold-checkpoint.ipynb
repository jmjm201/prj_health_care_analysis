```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import datetime
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import *
from tensorflow.keras.optimizers import Adam
import keras
from tensorflow.keras.layers import *
```

# 데이터 로드

```python
data = pd.read_csv('../datasets/prep_total.csv', index_col= 0)
```

# 데이터 전처리

```python
data = data[['검진 시 연령','체중', '허리둘레', 'BMI','수축기혈압', '이완기혈압']]
data.reset_index(inplace= True)
data.rename(columns={'고유값':'일련번호'}, inplace=True)
```

```python
data.head(20)
```

```python
data.columns
```

```python
data.dropna(axis=0, inplace= True)
```

# 데이터 그룹화

```python
grouped = data.groupby(['일련번호'])
grouped_df = {}
for key, group in grouped:
    grouped_df[key] = group
```

# 스케일링/시퀀스설정

```python
from sklearn.preprocessing import MinMaxScaler
```

```python
minmaxscaler = MinMaxScaler()
# Groupby한 것을 잘라서 각각 스케일링하고 시퀀스작업도 각각해야함
X_list = []
Y_list = []
for key in grouped_df.keys():
    # 스케일링
    scaled_data = minmaxscaler.fit_transform(grouped_df[key][['검진 시 연령', '체중',
                                                              '허리둘레', 'BMI', '수축기혈압', '이완기혈압']])
    
    # 시퀀스작업
    for i in range(0, len(scaled_data) - 3):
        _x = scaled_data[i : i + 3]
        _y = scaled_data[i + 3][5]
        X_list.append(_x)
        Y_list.append(_y)
```

```python
sequence_X = np.array(X_list)
sequence_Y = np.array(Y_list)
print(sequence_X.shape)
print(sequence_Y.shape)
```

# K-fold-cross-validation

```python
from sklearn.model_selection import KFold
```

```python
kf = KFold(n_splits=5)
kf.get_n_splits(sequence_X)
print(kf)
```

```python
for train_index, test_index in kf.split(sequence_X):
    print("TRAIN:", train_index, "TEST:", test_index)
    X_train, X_test = sequence_X[train_index], sequence_X[test_index]
    Y_train, Y_test = sequence_Y[train_index], sequence_Y[test_index]
```

```python
X_train.shape, X_test.shape, Y_train.shape
```

# Modeling1

```python
model_1 = Sequential()
model_1.add(LSTM(500, input_shape = (X_train.shape[1], X_train.shape[2]), activation= 'tanh'))
model_1.add(Flatten())
model_1.add(Dense(1))
model_1.compile(loss= 'mse', optimizer= 'adam')
model_1.summary()
```

```python
early_stop = keras.callbacks.EarlyStopping(monitor = 'val_loss', patience= 20)
fit_hist = model_1.fit(X_train, Y_train, epochs = 500, validation_data= (X_test, Y_test), shuffle= False, callbacks= [early_stop])
```
